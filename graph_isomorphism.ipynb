
# Graph Isomorphism Problem

### Summary

The Graph Isomorphism Problem is a problem for determining if a graph is structurally identical, or isomorphic, compared to another graph. 

An isomorphic graph is a graph that can have more than one form. If you compare two graphs that have the same number of nodes, edges and they have the same connectivity between edges, then even though they may visually appear different, they are essentially the same graph. [1]

An example of an isomorphic graph is shown below in the two python scripts. These create two graphs with the same number of points and (paying attention to the colour coding) have the same connections to each point making these two graphs isomorphic.


```python
import matplotlib.pyplot as plt

x1 = [1,1.5]
y1 = [3,4]
x2 = [1.5,2]
y2 = [4,3]
x3 = [1,1.25]
y3 = [3,2]
x4 = [1.25,1.75]
y4 = [2,2]
x5 = [1.75,2]
y5 = [2,3]
plt.plot(x1, y1, label = "Points : A,B Edge : (A,B)")
plt.plot(x2, y2, label = "Points : B,C Edge : (B,C)")
plt.plot(x5, y5, label = "Points : C,D Edge : (C,D)")
plt.plot(x4, y4, label = "Points : E,D Edge : (E,D)")
plt.plot(x3, y3, label = "Points : A,E Edge : (A,E)")

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Two lines on same graph!')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x1 = [3,4]
y1 = [3.25,3.25]
x2 = [4,3.15]
y2 = [3.25,2]
x3 = [3.15,3.5]
y3 = [2,4]
x4 = [3.5,3.85]
y4 = [4,2]
x5 = [3.85,3]
y5 = [2,3.25]
plt.plot(x1, y1, label = "Points : A,B Edge : (A,B)")
plt.plot(x2, y2, label = "Points : B,C Edge : (B,C)")
plt.plot(x3, y3, label = "Points : A,E Edge : (A,E)")
plt.plot(x4, y4, label = "Points : E,D Edge : (E,D)")
plt.plot(x5, y5, label = "Points : C,D Edge : (C,D)")

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Two lines on same graph!')
plt.legend()
plt.show()
```
