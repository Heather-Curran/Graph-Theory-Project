
# Graph Isomorphism Problem

## Introduction

The Graph Isomorphism Problem is a problem for determining if a graph is structurally identical, or isomorphic, compared to another graph. 

An isomorphic graph is a graph that can have more than one form. If you compare two graphs that have the same number of nodes, edges and they have the same connectivity between edges, then even though they may visually appear different, they are essentially the same graph. [1]

An example of an isomorphic graph is shown below in the two python scripts. These create two graphs with the same number of points and (paying attention to the colour coding) have the same connections to each point making these two graphs isomorphic.

### Graphs 3 & 4

The above two graphs are small enough that when a person looks at them, they may be able to tell that they are isomorphic. However, as the graphs increase is size, it quickly becomes very difficult if not impossible to tell if they are isomorphic just from looking at them.

There are four questions you can ask to determine if two graphs are isomorphic:

1. Are the number of nodes in the given graphs the same? 
2. Are the number of edges in the given graphs the same? 
3. Is the degree sequence in the given graphs the same? 
4. If the first graph is forming a cycle of length k with the help of vertices {v1, v2, v3, …. vk}, then another graph must also form the same cycle of the same length k with the help of vertices {v1, v2, v3, …. vk}. [2] 


We can answer these four questions in relation to the two more complicated graphs 3 and 4 below to discover if they are isomorphic or not.

1. Are the number of nodes in the given graphs the same?
    - Graph 3 has ten nodes.
    - Graph 4 has ten nodes.
        
  *Pass: The graphs have the same number of nodes.*


2. Are the number of edges in the given graphs the same? 
    - Graph 3 has thirteen edges. 
    - Graph 4 has thirteen edges.

 *Pass: The graphs have the same number of edges.*
  

3.	Is the degree sequence in the given graphs the same? 
    - Graph 3 has a degree sequence of (3,2,3,2,3,2,3,2,3,3)
    - Graph 4 has a degree sequence of (3,2,3,2,3,2,3,2,3,3)

 *Pass: The graphs have the same degree sequence.*


4.	If the first graph is forming a cycle of length k with the help of vertices {v1, v2, v3, …. vk}, then another graph must also form the same cycle of the same length k with the help of vertices {v1, v2, v3, …. vk}.  
    - Graph 3 forms a cycle of length 4 with the help of vertices {a,b,c,j,a}
    - Graph 4 forms a cycle of length 4 with the help of vertices {a,b,c,j,a}

### ***Conclusion: Graphs 3 and 4 are isomorphic.***

## Adjacency Matrix

Another way that we can check if two given graphs are isomorphic is if their adjacency matrices are the same. Two vertices, or nodes, are said to be adjacent if they are connected by an edge.

For Graph 7 below, its adjacency matrix would be:

      a b c d
    a 0 1 1 1 
    b 1 0 1 1
    c 1 1 0 0
    d 1 1 0 0
  
And for Graph 8 below, its adjacency matrix would also be:

      a b c d
    a 0 1 1 1
    b 1 0 1 1
    c 1 1 0 0
    d 1 1 0 0
  
As the two adjacency matrices are the same we can determine that the two graphs are isomorphic.

### ***Conclusion: Graphs 7 and 8 are isomorphic.***

## Computational Complexity

<br>

Within mathematics, computational complexity theory is a research area that focuses on classifying computational problems based on the resources required to solve said problems. In simple terms, it is a way of ranking problems based on their complexity. 

We know that there are algorithms that can solve whether two graphs are isomorphic or not. However, computer scientists have struggled to determine where in the classification of complexity the graph isomorphism question belongs. [3] 
The graph isomorphism problem is known to be NP (nondeterministic polynomial time), but it is not known if it belongs in either of its well-known subsets: P and NP-complete. [4] 

- P stands for Polynomial time and is a collection of decision problems, which are problems with a yes or no answer, that can be solved by a deterministic Turing machine in polynomial time. 

- NP is a collection of decision problems that can be solved by a non-deterministic Turing machine in polynomial time. P is a subset of NP as any problem that can be solved by a deterministic machine in polynomial time can be solved by a non-deterministic one.

- “An NP-hard problem is at least as hard as the hardest problem in NP, and it is the class of the problems such that every problem in NP reduces to NP-hard”.[5] 

- A problem is considered NP-complete if it is both NP and NP-hard.

## Babai's Algorithm

<br>

In 2015, mathematician Laszlo Babai presented a new algorithm for the graph isomorphism problem which he stated ruins in quasi-polynomial time. In 2017, he retracted this claim after an error was found. However, just five days later he announced that the error had been fixed and that the algorithm did run in quasi-polynomial time. This would put the graph isomorphism problem close to the P subset of NP. The fastest algorithm before this one had held is record for over thirty years.
